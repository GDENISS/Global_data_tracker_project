```python
columns_to_drop = []

for column in filtered_data.columns:
    if filtered_data[column].isnull().sum() == 790:
        columns_to_drop.append(column)
```

```python
new_data = filtered_data.drop(columns = columns_to_drop)
```

```python
new_data.head()
```

```python
new_data.isnull().sum()
```

```python
empty = []
for column in new_data.columns:
    if new_data[column].isnull().sum() >= 50:
        empty.append(column)
```

```python
final_data = new_data.drop(columns = empty )
```
